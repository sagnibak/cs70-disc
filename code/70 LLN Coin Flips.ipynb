# Demo of the Law of Large Numbers using coins flips and die rolls

```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib as mpl
mpl.rcParams["figure.figsize"] = [8, 6]
mpl.rcParams["font.size"] = 14


MIN_FLIPS = 10
MAX_FLIPS = 10000
STEP_SIZE = 5

def flip_coin(x):
    """Flip a fair coin `x` times and return the average number Heads."""
    return np.mean(np.random.randint(0, 2, (x,)))

def roll_die(x):
    """Roll a fair die `x` times and return the average rolled value"""
    return np.mean(np.random.randint(1, 7, (x,)))

num_reps = list(range(MIN_FLIPS, MAX_FLIPS + 1, STEP_SIZE))
```

## Coin flips

Let's flip a coin x times and record the percentage (also the average number) of Heads.
What do you think the shape of a graph of: 

Number of independent coin flips (x axis) vs Average number of heads (y axis) looks like?

```python
coin_mean = [0.5] * len(num_reps)

plt.ylim((0.2, 0.8)) 
plt.plot(num_reps, coin_mean, 'b')

plt.title('# independent coin flips vs Average number of heads')

plt.xlabel('Number of coin flips')
plt.ylabel('Average number of heads');
```

### Let's plot the graph:

HHHTTH -> 4/6 = 0.667
THTTHHHTT -> 4/9 = 0.444

```python
coin_averages = [flip_coin(n) for n in num_reps]
mean = [0.5 for _ in num_reps]

plt.ylim((0.2, 0.8)) 
plt.plot(num_reps, coin_averages, 'r')
plt.plot(num_reps, mean, 'b')

plt.title('# independent coin flips vs Average number of heads')

plt.xlabel('Number of coin flips')
plt.ylabel('Average number of heads');
```

## Die rolls

Let's roll a die k times and record the average of the k rolled values.
What do you think the shape of a graph of: 

Number of independent die rolls (x axis) vs average of the k rolled values (y axis) looks like?

```python
coin_mean = [3.5] * len(num_reps)

plt.ylim((2.5, 4.5)) 
plt.plot(num_reps, coin_mean, 'b')

plt.title('# independent die rolls vs Average value of rolls')

plt.xlabel('Number of dice rolls')
plt.ylabel('Average of rolled values');
```

### Let's plot the graph:

```python
die_averages = [roll_die(n) for n in range(MIN_FLIPS, MAX_FLIPS + 1, STEP_SIZE)]
die_mean = [3.5 for _ in num_reps]

plt.ylim((2.5, 4.5)) 
plt.plot(num_reps, die_averages, 'r')
plt.plot(num_reps, die_mean, 'b')

plt.title('# independent die rolls vs Average value of rolls')

plt.xlabel('Number of dice rolls')
plt.ylabel('Average of rolled values');
```
